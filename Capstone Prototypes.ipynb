```python
import matplotlib.pyplot as plt
from scipy.io import wavfile as wav
from scipy.fftpack import fft
import numpy as np
```

```python
rate, song = wav.read('XC178578 - European Robin - Erithacus rubecula- (online-audio-converter.com).wav')
fft_out = fft(song)

clean = np.abs(fft_out)

comb = []

for i in clean:
    comb.append(sum(i))

plt.plot(comb, color = 'orange')
plt.show()
```

```python
rate, alarm = wav.read('XC94909 - European Robin - Erithacus rubecula- (online-audio-converter.com).wav')
fft_out1 = fft(alarm)

clean1 = np.abs(fft_out1)

comb1 = []

for i in clean1:
    comb1.append(sum(i))

plt.plot(comb1)
plt.show()
```

```python
print(max(comb), max(comb1))
print(min(comb), min(comb1))
print("The avg frequency of the song is %.2f" %np.mean(comb))
print("The avg frequency of the alarm is %.2f" %np.mean(comb1))
print("The median frequency of the song is %.2f" %np.median(comb))
print("The median frequency of the alarm is %.2f" %np.median(comb1))
```
